```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
trainable = np.load('trainable_stft.npy')
nontrainable = np.load('nontrainable_stft.npy')

trainable2 = np.load('trainable_mel.npy')
nontrainable2 = np.load('nontrainable_mel.npy')

trainable3 = np.load('trainable_cqt.npy')
nontrainable3 = np.load('nontrainable_cqt.npy')

trainable_lin = np.load('trainable_stft_linear.npy')
nontrainable_lin = np.load('nontrainable_stft_linear.npy')

trainable2_lin = np.load('trainable_mel_lin.npy')
nontrainable2_lin = np.load('nontrainable_mel_lin.npy')

trainable3_lin = np.load('trainable_cqt_lin.npy')
nontrainable3_lin = np.load('nontrainable_cqt_lin.npy')
```

```python
fig, ax = plt.subplots(3,2,figsize=(12,15))
cols = ['Linear', 'CNN']
rows = ['STFT', 'MelSpec', 'CQT']

ax[0,0].plot(trainable_lin)
ax[0,0].plot(nontrainable_lin)
ax[0,0].set_yscale('log')
ax[0,0].set_title('Linear', size=18)
ax[0,0].legend(['Trainable', 'Non-Trainable'])
ax[0,0].set_ylabel('STFT', size=18)
ax[0,0].tick_params(labelsize=14)
ax[0,0].set_ylim(1e-6,0.2)

ax[0,1].plot(trainable)
ax[0,1].plot(nontrainable)
ax[0,1].set_yscale('log')
ax[0,1].set_title('CNN', size=18)
ax[0,1].legend(['Trainable', 'Non-Trainable'])
ax[0,1].tick_params(labelsize=14)
ax[0,1].set_ylim(1e-6,0.2)

ax[1,0].plot(trainable2_lin)
ax[1,0].plot(nontrainable2_lin)
ax[1,0].set_yscale('log')
ax[1,0].legend(['Trainable', 'Non-Trainable'])
ax[1,0].set_ylabel('MelSpec', size=18)
ax[1,0].set_xlabel('Epoch', size=14)
ax[1,0].tick_params(labelsize=14)
ax[1,0].set_ylim(1e-6,0.2)

ax[1,1].plot(trainable2)
ax[1,1].plot(nontrainable2)
ax[1,1].set_yscale('log')
ax[1,1].legend(['Trainable', 'Non-Trainable'])
ax[1,1].set_xlabel('Epoch', size=14)
ax[1,1].tick_params(labelsize=14)
ax[1,1].set_ylim(1e-6,0.2)

ax[2,0].plot(trainable3_lin)
ax[2,0].plot(nontrainable3_lin)
ax[2,0].set_yscale('log')
ax[2,0].legend(['Trainable', 'Non-Trainable'])
ax[2,0].set_ylabel('CQT', size=18)
ax[2,0].set_xlabel('Epoch', size=14)
ax[2,0].tick_params(labelsize=14)
ax[2,0].set_ylim(1e-6,0.2)

ax[2,1].plot(trainable3)
ax[2,1].plot(nontrainable3)
ax[2,1].set_yscale('log')
ax[2,1].legend(['Trainable', 'Non-Trainable'])
ax[2,1].set_xlabel('Epoch', size=14)
ax[2,1].tick_params(labelsize=14)
ax[2,1].set_ylim(1e-6,0.2)

for ax_idx, col in zip(ax[0], cols):
    ax_idx.set_title(col, size=18)
    
# for ax_idx, row in zip(ax[:,0], rows):
#     ax_idx.annotate(row, xy=(0, 0), xytext=(ax_idx.yaxis.labelpad-5, ax_idx.xaxis.labelpad-3), rotation=90,
#                 xycoords=ax_idx.yaxis.label, size=18)
```
