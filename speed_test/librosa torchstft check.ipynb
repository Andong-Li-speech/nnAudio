```python
import torch
from librosa import stft
import numpy as np
import matplotlib.pyplot as plt
```

```python
# t = 2*np.pi*np.linspace(0,1,5000)
# x = np.sin(50*t)
x = np.random.randn(10,10000).astype(np.float32)
lib_output = stft(x[0], n_fft=2048, hop_length=512)
window = torch.hann_window(2048)
```

```python
x = torch.nn.Parameter(x)
```

```python
torch_output = torch.stft(x, n_fft=2048, hop_length=512, window=window)
```

```python
plt.imshow(torch.sqrt(torch_output[0].pow(2).sum(-1)), aspect='auto')
```

```python
plt.imshow(abs(lib_output), aspect='auto')
```

```python
np.allclose(torch.sqrt(torch_output[0].pow(2).sum(-1)), abs(lib_output), atol=1e-5, rtol=1e-6)
```
