```python
import numpy as np
from librosa import stft
import matplotlib.pyplot as plt
import time
import tqdm
import pandas as pd
```

```python
y_list = np.load('../y_list.npy')
```

```python
spec_list = []
timing = []
for e in range(5):
    t_start = time.time()
    for i in tqdm.tqdm(y_list, leave=True):
        spec = stft(i)
        spec_list.append(abs(spec))
    time_used = time.time()-t_start
    print(time_used)
    timing.append(time_used)
```

```python
print("mean = ",np.mean(timing))
print("std = ", np.std(timing))
```

```python
data = pd.DataFrame(timing,columns=['t_avg'])
data['Type'] = 'librosa'
```

```python
data.to_csv('librosa_Spec')
```
